```python
def list_files_within_date_range(start_date, end_date, s3_bucket, s3_prefix):
    date_list = []
    current_date = start_date

    # Loop over each day in the date range
    while current_date <= end_date:
        year, month, day = current_date.year, current_date.month, current_date.day
        date_prefix = f"{s3_prefix}/year={year}/month={str(month).zfill(2)}/day={str(day)}/"
        print("date_prefix",date_prefix)
        # Check if files exist for the date
        response = s3.list_objects_v2(Bucket=s3_bucket, Prefix=date_prefix)
        
        # If files are found for the date, it means they are already processed
        if 'Contents' not in response:
            # Only add dates with no files in S3 (i.e., unprocessed dates)
            date_list.append(current_date)
        
        current_date += timedelta(days=1)
    
    return date_list
```

```python
def list_files_within_date_range(start_date, end_date, s3_bucket, s3_prefix):
    date_list = []
    end_date_files = []
    previous_30_days_files = []
    current_date = start_date

    # Loop over each day in the date range
    while current_date <= end_date:
        year, month, day = current_date.year, current_date.month, current_date.day
        date_prefix = f"{s3_prefix}/year={year}/month={str(month).zfill(2)}/day={str(day)}/"
        print("date_prefix", date_prefix)
        
        # Check if files exist for the date
        response = s3.list_objects_v2(Bucket=s3_bucket, Prefix=date_prefix)
        
        # If no files are found for the date, mark it as unprocessed
        if 'Contents' not in response:
            date_list.append(current_date)
        else:
            # Filter for .csv files only and sort them by LastModified timestamp
            csv_files = [obj for obj in response['Contents'] if obj['Key'].endswith('.csv')]
            
            if csv_files:
                # Sort the .csv files by LastModified timestamp (descending)
                sorted_csv_files = sorted(csv_files, key=lambda obj: obj['LastModified'], reverse=True)
                
                # Get the latest .csv file (the first in the sorted list)
                latest_csv_file = f"s3://{s3_bucket}/{sorted_csv_files[0]['Key']}"
                
                # Add the latest .csv file to the appropriate list
                if current_date == end_date:
                    end_date_files.append(latest_csv_file)
                else:
                    previous_30_days_files.append(latest_csv_file)
        
        current_date += timedelta(days=1)
    
    return date_list, end_date_files, previous_30_days_files
```
